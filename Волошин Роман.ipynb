```python
Погані_люди = ["Дедліпаркур Владислав Гарболінський"]
print("Введіть імена людей, яких ви вважаєте поганими (відокремлюйте імена комами а також не пишіть ім'я Роман):")
імена_поганих = input().split(",")
погані_люди.extend(імена_поганих)
всі_люди = ["Мар'яна", "Роман чений хороший хлопчик", "Максим інколи поганий моментами", "", ""]
хороші_люди = [особа for особа in всі_люди if особа not in погані_люди]
print("Погані люди:", погані_люди)
print("Хороші люди:", хороші_люди)
```

# Новий розділ

```python
import matplotlib.pyplot as plt
import numpy as np
a = 1
b = 2
c = 1
x = np.linspace(-10, 10, 100)
y = a * x**2 + b * x + c
plt.plot(x, y, label='Парабола: $y = ax^2 + bx + c$')

plt.title('Графік параболи')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)
plt.show()
```
